```python
import tkinter as tk
from tkinter import Text, Entry, Button, Scrollbar
import random
from rapidfuzz import fuzz

# Function to get the best matching response
def get_response(user_input):
    best_match = None
    best_score = 0
    
    # Iterate through each pair of key phrases and responses
    for key_phrases, responses in pairs:
        # Fuzzy match the user's input against each key phrase
        for key_phrase in key_phrases:
            score = fuzz.ratio(user_input.lower(), key_phrase.lower())
            if score > best_score:
                best_score = score
                best_match = responses
    
    # Set a threshold score for accepting a match (e.g., 75)
    if best_score >= 75 and best_match:
        return random.choice(best_match)
    else:
        return "I'm sorry, I don't understand that question. Can you please rephrase?"

# Create the main application window
root = tk.Tk()
root.title("Chatbot")
root.geometry("500x400")
root.configure(bg="#2C3E50")

# Function to send messages to the chatbot
def send_message():
    user_message = user_input.get()
    user_input.delete(0, tk.END)
    chat_history.insert(tk.END, "You: " + user_message + "\n", 'user')
    response = get_response(user_message)
    chat_history.insert(tk.END, "Chatbot: " + response + "\n", 'bot')
    chat_history.yview(tk.END)

# Chat history Text widget
chat_history = Text(root, width=50, height=15, bg="#ECF0F1", fg="#2C3E50", font=("Helvetica", 12), padx=10, pady=10, wrap='word')
chat_history.pack(padx=10, pady=10)
scrollbar = Scrollbar(root)
scrollbar.pack(side=tk.RIGHT, fill=tk.Y)
chat_history.config(yscrollcommand=scrollbar.set)
scrollbar.config(command=chat_history.yview)

# User input Entry widget
user_input = Entry(root, width=40, bg="#ECF0F1", fg="#2C3E50", font=("Helvetica", 12))
user_input.pack(padx=10, pady=5, side=tk.LEFT, expand=True)

# Send button
send_button = Button(root, text="Send", command=send_message, bg="#3498DB", fg="#ECF0F1", font=("Helvetica", 12, "bold"))
send_button.pack(padx=10, pady=5, side=tk.RIGHT)

# Tag configuration for styling chat history text
chat_history.tag_config('user', foreground="#2980B9")
chat_history.tag_config('bot', foreground="#27AE60")

# Start the Tkinter main loop
root.mainloop()
```
